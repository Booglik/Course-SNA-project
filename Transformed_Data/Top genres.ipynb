```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

hr = pd.read_csv('HR_genres.csv', index_col='Unnamed: 0')
hu = pd.read_csv('HU_genres.csv', index_col='Unnamed: 0')
ro = pd.read_csv('RO_genres.csv', index_col='Unnamed: 0')
```

```python
datasets = {'hr': hr, 'hu': hu, 'ro': ro}
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, (country, data) in enumerate(datasets.items()):
    sum_country = pd.DataFrame(data.sum(), columns=['sum'])
    
    total_sum_country = sum_country['sum'].sum()
    
    sum_country['share'] = sum_country['sum'] / total_sum_country * 100
    
    sum_country_sorted = sum_country.sort_values(by='share', ascending=False)
    
    top_genres_country = sum_country_sorted.head(10).sort_values(by='share', ascending=True)
    
    colors = sns.color_palette('viridis', len(top_genres_country))
    
    axs[i].barh(top_genres_country.index, top_genres_country['share'], color=colors)
    axs[i].set_xlabel('Share (%)')
    axs[i].set_ylabel('Genre')
    axs[i].set_title('Top 10 Genres by Share - {}'.format(country.upper()))

plt.tight_layout()

plt.savefig('subplots.png', dpi=300)
plt.show()
```
